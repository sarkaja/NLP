```python
import spacy
import pandas as pd
import numpy as np
import spacy.cli
import seaborn as sns
spacy.cli.download("en_core_web_md")
nlp = spacy.load("en_core_web_md")

import nltk
import string
import fasttext
import contractions
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer


```

```python
drew = pd.read_csv("reviews.csv", 
                 index_col=False,
                 low_memory=False)
drew
```

```python
dus = pd.read_csv("users.csv", 
                 index_col=False,
                 low_memory=False)
dus
```

```python
dbus = pd.read_csv("businesses.csv", 
                 index_col=False,
                 low_memory=False)
dbus
```

```python
#Spojení datasetu reviews a businesses a vytvoření "mezi datasetu"
dmez = pd.merge(drew, dbus, left_on='business_id', right_on='business_id', how='left')
dmez.head(6)
```

```python
#Spojení mezi datasetu a datasetu 'users' do jednoho dfin
dfin = pd.merge(dmez, dus, left_on='user_id', right_on='user_id', how='left')
dfin.head(6)
```

```python
dfin.columns
```

```python
dfin['date'] = pd.to_datetime(dfin.date)
```

```python
dfin = dfin.drop(['hours.Monday','postal_code','hours.Thursday','hours.Tuesday','hours.Friday','hours.Wednesday',
             'hours.Saturday','hours.Sunday','longitude','latitude'], axis=1)
```

```python
dfin
```

```python
filtr = [col for col in dfin if col.startswith('attribut')]
filtr
```

```python
dfdal = dfin[filtr]
dfdal
```

```python
#Jaké jednotlivé atributy jsou nulové 
for col in dfin.columns:
    print(col, dfin[col].isnull().sum())
```

```python
for column in dfdal:
    dfdal[column].fillna(False, inplace=True)
dfdal
```

```python
for column in dfin[filtr]:
    dfin[column].replace('no', False)

dfin
```

```python
#Odstranění zkrácených tvarů
dfin['text_upr'] = dfin['text'].apply(lambda x: [contractions.fix(word) for word in x.split()])
dfin.head()
```

```python
#Převod sloupce upraveného textu na string
dfin['text_upr_str'] = [' '.join(map(str, l)) for l in dfin['text_upr']]
dfin.head()
```

```python
#Tokenizace
dfin['tokenized'] = dfin['text_upr_str'].apply(word_tokenize)
dfin.head()
```

```python
#Odstranění velkých písmen
dfin['lower'] = dfin['tokenized'].apply(lambda x: [word.lower() for word in x])
dfin.head()
```

```python
#Odstranění interpunkce
punc = string.punctuation
dfin['no_punc'] = dfin['lower'].apply(lambda x: [word for word in x if word not in punc])
dfin.head()
```

```python
#Odstranění neutrálních slov
stop_words = set(stopwords.words('english'))
dfin['stopwords_removed'] = dfin['no_punc'].apply(lambda x: [word for word in x if word not in stop_words])
dfin.head()
```

```python
#Určení slovních druhů slov v textu
dfin['pos_tags'] = dfin['stopwords_removed'].apply(nltk.tag.pos_tag)
dfin.head()
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
dfin['wordnet_pos'] = dfin['pos_tags'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
dfin.head()
```

```python
wnl = WordNetLemmatizer()
dfin['lemmatized'] = dfin['wordnet_pos'].apply(lambda x: [wnl.lemmatize(word, tag) for word, tag in x])
dfin.head()
```

```python
dfin.to_csv('dfin_clean.csv')
```
